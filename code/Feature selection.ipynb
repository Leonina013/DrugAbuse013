```python
import pandas as pd
import numpy as np
```

```python
file = '..//data//nsduh_with_final_column.csv'
main_data = pd.read_csv(file, sep=',')
```

```python
x = main_data.drop('CATEGORY',axis=1)
y = main_data['CATEGORY']
```

### Standarization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x)
scaled_features = scaler.transform(x)
```

```python
scaled_features
```

### PCA extraction

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.90)
reduced = pca.fit_transform(x)
```

```python
reduced_data = pd.DataFrame(reduced)
reduced_data.head()
```

```python
reduced_data.info()
```

```python
coverage = np.sum(pca.explained_variance_ratio_)
coverage
```

```python
reduced_data['CATEGORY'] = y
reduced_data.head()
```

```python
reduced_data.to_csv('..\\data\\nsduh_reduced.csv',index=False)
```

## TPOT to choose pipeline

```python
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
```

```python
x = reduced_data.drop('CATEGORY',axis=1)
y = reduced_data['CATEGORY']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.7, random_state=121)
```

```python
pipeline_optimizer = TPOTClassifier(generations=5, population_size=20, cv=5,
                                    random_state=42, verbosity=2, max_time_mins= 5, periodic_checkpoint_folder= 'D:\Dokumenty\analityka\magisterka\code')
pipeline_optimizer.fit(x_train, y_train)
print(pipeline_optimizer.score(x_test, y_test))
```

```python
pipeline_optimizer.export('tpot_exported_pipeline.py')
```
