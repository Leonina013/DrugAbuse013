```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split
```

```python
x = reduced_data.drop('RISKGR',axis=1)
y = reduced_data['RISKGR']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=121)
```

```python
selector = SelectFromModel(RandomForestClassifier(n_estimators = 100))
selector.fit(x_train, y_train)
```

```python
selector.get_support()
```

```python
selected_features= x_train.columns[(selector.get_support())]
selected_features
```

```python
selected_data = pd.DataFrame(reduced_data, columns=selected_features)
selected_data.head()
```

```python
selected_data['RISKGR'] = y
selected_data.head()
```

```python
selected_data.to_csv('..\\data\\nsduh_selected.csv',index=False)
```

## TPOT to choose pipeline

```python
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
```

```python
x = selected_data.drop('RISKGR',axis=1)
y = selected_data['RISKGR']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.7, random_state=121)
```

```python
pipeline_optimizer = TPOTClassifier(population_size=20, cv=5, random_state=42, 
                                    verbosity=2, early_stop = 7)
pipeline_optimizer.fit(x_train, y_train)
print(pipeline_optimizer.score(x_test, y_test))
pipeline_optimizer.export('tpot_exported_pipeline.py')
```
